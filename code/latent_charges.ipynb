# Latent Charges

Here we explore the results of the latent charges model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../dl_paper/SourceData_Figure_1.csv')
```

```python
path_pref = "/fs/pool/pool-cox-data08/Juan/ccs/Data/second_split/"
df_cc = pd.read_csv(f"{path_pref}charged_corrected.csv")
```

```python
df1['m/t'] = (df1['Mass'] + df_cc['t'])/df_cc['t']
```

# Plots

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df1['m/t'], df1['CCS'], s=0.1, c = df_cc['t'])
ax.set_title('CCS vs m/t')
ax.set_xlabel('m/t')
ax.set_ylabel('CCS')
```
